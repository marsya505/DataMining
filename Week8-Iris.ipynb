```python
# Import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from google.colab import drive
drive.mount('/content/drive')

# Load & Preprocess Data Iris
df = pd.read_csv('/content/drive/MyDrive/Kuliah/DataMining/Dataset/Iris/Iris.csv')
X_iris = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]

# Standardisasi
scaler_iris = StandardScaler()
X_iris_scaled = scaler_iris.fit_transform(X_iris)

# K-Means (K=3)
kmeans_iris = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans_iris_labels = kmeans_iris.fit_predict(X_iris_scaled)

# DBSCAN
dbscan_iris = DBSCAN(eps=0.8, min_samples=5)
dbscan_iris_labels = dbscan_iris.fit_predict(X_iris_scaled)

# Agglomerative (K=3)
agg_iris = AgglomerativeClustering(n_clusters=3)
agg_iris_labels = agg_iris.fit_predict(X_iris_scaled)

# Visualisasi dengan PCA
pca_iris = PCA(n_components=2)
iris_2d = pca_iris.fit_transform(X_iris_scaled)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Plot K-Means
axes[0].scatter(iris_2d[:, 0], iris_2d[:, 1], c=kmeans_iris_labels, cmap='viridis', edgecolor='k')
axes[0].set_title('K-Means: Data Iris')
axes[0].set_xlabel('PCA 1')
axes[0].set_ylabel('PCA 2')

# Plot DBSCAN
axes[1].scatter(iris_2d[:, 0], iris_2d[:, 1], c=dbscan_iris_labels, cmap='viridis', edgecolor='k')
axes[1].set_title('DBSCAN: Data Iris (Ungu/ -1 = Noise)')
axes[1].set_xlabel('PCA 1')

# Plot Agglomerative
axes[2].scatter(iris_2d[:, 0], iris_2d[:, 1], c=agg_iris_labels, cmap='viridis', edgecolor='k')
axes[2].set_title('Agglomerative: Data Iris')
axes[2].set_xlabel('PCA 1')

plt.tight_layout()
plt.show()

# Print DBSCAN Info
print("Cluster yang terbentuk (-1 adalah noise):", np.unique(dbscan_iris_labels))
print(pd.Series(dbscan_iris_labels).value_counts())
```
