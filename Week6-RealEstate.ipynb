```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

from google.colab import files
uploaded = files.upload()

data = pd.read_csv("RealEstate.csv")
data.head()

X_re = data.drop(['No', 'Y house price of unit area'], axis=1)
y_re = data['Y house price of unit area']

X_train_re, X_test_re, y_train_re, y_test_re = train_test_split(
    X_re, y_re, test_size=0.2, random_state=42
    )

model_re = LinearRegression()
model_re.fit(X_train_re, y_train_re)

y_pred_re = model_re.predict(X_test_re)
print("R-squared (R2):", r2_score(y_test_re, y_pred_re))
print("Koefisien Jarak MRT:", model_re.coef_[2])

plt.figure(figsize=(8, 6))
plt.scatter(y_test_re, y_pred_re, alpha=0.6, color='blue')

plt.plot(
    [y_test_re.min(), y_test_re.max()],
     [y_test_re.min(), y_test_re.max()],
    color='red', linewidth=2)

plt.title('Real Estate: Harga Asli vs Prediksi Model')
plt.xlabel('Harga Asli (Actual Price)')
plt.ylabel('Harga Prediksi (Predicted Price)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
