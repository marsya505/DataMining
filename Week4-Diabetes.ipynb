```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (asumsi dataset sudah di-upload)
df = pd.read_csv('diabetes.csv')

# List kolom yang tidak boleh bernilai 0
cols_with_zeros = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

# Ganti 0 dengan NaN agar mudah diproses
for col in cols_with_zeros:
    df[col] = df[col].replace(0, np.nan)

# Imputasi NaN dengan Median berdasarkan Outcome (0 atau 1)
# Ini lebih akurat daripada menggunakan median global
for col in cols_with_zeros:
    df[col] = df.fillna(df.groupby('Outcome')[col].transform('median'))

print("Jumlah Missing Values setelah perbaikan:\n", df.isnull().sum())
```
