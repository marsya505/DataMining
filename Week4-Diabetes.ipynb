```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import drive
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/MyDrive/Kuliah/DataMining/Dataset/Diabetes/diabetes.csv')

# DATA CLEANING
#glukosa, tekanan darah, dll tidak mungkin nol bagi orang hidup
cols_with_zeros = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

#ganti 0 menjadi NaN agar tidak merusak perhitungan statistik
for col in cols_with_zeros:
    df[col] = df[col].replace(0, np.nan)

#gunakan transform('median') agar nilai yang kosong diisi median kelompoknya masing-masing
for col in cols_with_zeros:
    median_per_group = df.groupby('Outcome')[col].transform('median')
    df[col] = df[col].fillna(median_per_group)

print("Pembersihan selesai. Jumlah missing values:\n", df.isnull().sum())
print("-" * 30)

# VISUALISASI DESKRIPTIF
# Distribusi Target (Outcome)
plt.figure(figsize=(6, 4))
sns.countplot(x='Outcome', data=df, palette='viridis')
plt.title('Proporsi Pasien: Sehat (0) vs Diabetes (1)')
plt.show()

# Heatmap Korelasi
#untuk melihat fitur mana yang paling kuat hubungannya dengan diabetes
plt.figure(figsize=(10, 8))
correlation = df.corr()
mask = np.triu(np.ones_like(correlation, dtype=bool))
sns.heatmap(correlation, annot=True, fmt='.2f', mask=mask, cmap='coolwarm', linewidths=0.5)
plt.title('Korelasi Antar Fitur Medis')
plt.show()

#distribusi Glukosa & BMI (Fitur Paling Berpengaruh)
fig, ax = plt.subplots(1, 2, figsize=(15, 6))

sns.histplot(data=df, x='Glucose', hue='Outcome', kde=True, ax=ax[0], palette='magma')
ax[0].set_title('Distribusi Glukosa berdasarkan Status Diabetes')

sns.violinplot(x='Outcome', y='BMI', data=df, ax=ax[1], palette='Set2', split=True)
ax[1].set_title('Sebaran BMI terhadap Kelompok Pasien')

plt.show()

# Pairplot untuk Melihat Pemisahan Data
#memvisualisasikan hubungan antar fitur utama
top_features = ['Glucose', 'BMI', 'Age', 'Outcome']
sns.pairplot(df[top_features], hue='Outcome', palette='husl', diag_kind='kde')
plt.suptitle('Hubungan Antar Variabel Kunci', y=1.02)
plt.show()
```
