```python
!pip install mlxtend

import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

from google.colab import drive
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/MyDrive/Kuliah/DataMining/Dataset/Chapter05Dataset/Chapter05DataSet.csv')

cols = ['Family', 'Hobbies', 'Social_Club', 'Political', 'Professional', 'Religious', 'Support_Group']
data = df[cols]

data = data.astype(bool)

frequent_items = apriori(data, min_support=0.1, use_colnames=True)
frequent_items

rules = association_rules(frequent_items, metric="confidence", min_threshold=0.5)

rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]

import matplotlib.pyplot as plt

plt.scatter(rules['support'], rules['confidence'])
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.show()
```
