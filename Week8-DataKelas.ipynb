```python
# Import library
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from google.colab import drive
drive.mount('/content/drive')

df = pd.read_csv('/content/drive/MyDrive/Kuliah/DataMining/Dataset/DataKelas/datakelas.csv')
# Ambil fitur numerik untuk clustering
X_kelas = df[['GPA', 'Height', 'Weight', 'Fuel/month']]

# Standardisasi
scaler_kelas = StandardScaler()
X_kelas_scaled = scaler_kelas.fit_transform(X_kelas)

# K-MEANS MODEL
# Buat jadi 3 cluster
kmeans_kelas = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans_kelas_labels = kmeans_kelas.fit_predict(X_kelas_scaled)

# Visualisasi PCA
pca_kelas = PCA(n_components=2)
kelas_2d = pca_kelas.fit_transform(X_kelas_scaled)

plt.figure(figsize=(8, 6))
plt.scatter(kelas_2d[:, 0], kelas_2d[:, 1], c=kmeans_kelas_labels, cmap='plasma', edgecolor='k')
plt.title('K-Means Clustering: Data Kelas (PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
