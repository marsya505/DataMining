```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from sklearn.datasets import load_iris

# Load dataset Iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target
print(df.head())

# Standarisasi data (semua kolom termasuk species sebagai fitur)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

# Elbow Method untuk menentukan k optimal
inertia = []
K = range(1, 10)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K, inertia, 'bx-')
plt.xlabel('Jumlah Cluster (k)')
plt.ylabel('Inertia')
plt.title('Metode Elbow untuk Menentukan k - Iris Dataset')
plt.grid(True)
plt.show()

# K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
km_labels = kmeans.fit_predict(X_scaled)
sil_kmeans = silhouette_score(X_scaled, km_labels)

# DBSCAN Clustering
dbscan = DBSCAN(eps=0.6, min_samples=5)
db_labels = dbscan.fit_predict(X_scaled)
mask = db_labels != -1
sil_dbscan = silhouette_score(X_scaled[mask], db_labels[mask])

# Agglomerative Clustering
agg_clust = AgglomerativeClustering(n_clusters=3)
agg_labels = agg_clust.fit_predict(X_scaled)
sil_agg = silhouette_score(X_scaled, agg_labels)

# Tampilkan hasil perbandingan
print(f"\nSilhouette Score K-Means      : {sil_kmeans:.3f}")
print(f"Silhouette Score DBSCAN       : {sil_dbscan:.3f}")
print(f"Silhouette Score Agglomerative: {sil_agg:.3f}")

# Visualisasi perbandingan ketiga metode
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

axes[0].scatter(df['petal length (cm)'], df['petal width (cm)'], c=km_labels, cmap='viridis', s=30)
axes[0].set_title(f'K-Means (k=3)\nSilhouette Score: {sil_kmeans:.3f}')
axes[0].set_xlabel('Petal Length (cm)')
axes[0].set_ylabel('Petal Width (cm)')
axes[0].grid(True)

axes[1].scatter(df['petal length (cm)'], df['petal width (cm)'], c=db_labels, cmap='viridis', s=30)
axes[1].set_title(f'DBSCAN (eps=0.6, min_samples=5)\nSilhouette Score: {sil_dbscan:.3f}')
axes[1].set_xlabel('Petal Length (cm)')
axes[1].set_ylabel('Petal Width (cm)')
axes[1].grid(True)

axes[2].scatter(df['petal length (cm)'], df['petal width (cm)'], c=agg_labels, cmap='viridis', s=30)
axes[2].set_title(f'Agglomerative (k=3)\nSilhouette Score: {sil_agg:.3f}')
axes[2].set_xlabel('Petal Length (cm)')
axes[2].set_ylabel('Petal Width (cm)')
axes[2].grid(True)

plt.suptitle('Perbandingan Metode Clustering pada Dataset Iris', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

# Bar chart perbandingan Silhouette Score
methods = ['K-Means\n(k=3)', 'DBSCAN\n(eps=0.6, min=5)', 'Agglomerative\n(k=3)']
scores  = [sil_kmeans, sil_dbscan, sil_agg]

plt.figure(figsize=(8, 5))
plt.bar(methods, scores, color=['#2196F3', '#FF5722', '#4CAF50'])
plt.ylabel('Silhouette Score')
plt.title('Perbandingan Silhouette Score - Iris Dataset')
plt.ylim(0, 1)
plt.grid(True, axis='y')
for i, v in enumerate(scores):
    plt.text(i, v + 0.01, f'{v:.3f}', ha='center', fontweight='bold')
plt.show()
```
