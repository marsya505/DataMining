```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

from google.colab import files
uploaded = files.upload()

data = pd.read_csv("insurance.csv")
data.head()

data = pd.get_dummies(data, drop_first=True)

X_ins = data.drop('charges', axis=1)
y_ins = data['charges']

X_train_ins, X_test_ins, y_train_ins, y_test_ins = train_test_split(
    X_ins, y_ins, test_size=0.2, random_state=42
    )

model_ins = LinearRegression()
model_ins.fit(X_train_ins, y_train_ins)

y_pred_ins = model_ins.predict(X_test_ins)
print("R-squared (R2):", r2_score(y_test_ins, y_pred_ins))

plt.figure(figsize=(8, 6))
plt.scatter(y_test_ins, y_pred_ins, alpha=0.6, color='green')

plt.plot(
    [y_test_ins.min(), y_test_ins.max()],
     [y_test_ins.min(), y_test_ins.max()],
    color='red', linewidth=2)

plt.title('Insurance: Biaya Medis Asli vs Prediksi Model')
plt.xlabel('Biaya Medis Asli (Actual Charges)')
plt.ylabel('Biaya Prediksi (Predicted Charges)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
